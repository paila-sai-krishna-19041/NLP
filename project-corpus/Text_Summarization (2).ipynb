Lematization

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
def lemetize(corpus4):
    word_tokens = word_tokenize(corpus4)
    filtered_word = [lem.lemmatize(word) for word in word_tokens]
    filtered_word = " ".join(filtered_word)
    return filtered_word
print(lemetize(corpus4))
```

```python
#POS TAGGING:
from collections import Counter
import nltk
#text = "Sukanya, Rajib and Naba are my good friends. " 
c=Counter()
tag=[]
for text in df:

#text="मुझेिदन तो याद नहींिकुमहीना िसतर यानी आिन  का था। शाम सेही बादल छाने"
  lower_case=text.lower()
  tokens=nltk.word_tokenize(lower_case)
  tags=nltk.pos_tag(tokens)
  counts=Counter(tag for word, tag in tags)
  c.update(counts)
  print(c,end="\n\n")
  tag.append(tags)
  print(tags)
```

# N-Gram analysis


```python


def generate_ngrams(text, n_gram=1, stop=True):
    """
    Simple n-gram generator.
    """
    stop = set(stopwords) if stop else {}

    token = [
        token for token in text.lower().split(" ") if token != "" if token not in stop
    ]
    z = zip(*[token[i:] for i in range(n_gram)])
    ngrams = [" ".join(ngram) for ngram in z]

    return ngrams
```

```python
def CountFrequency(my_list):
    freq = {}
    for item in my_list:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
    return freq
```

Unigram analysis

```python
from nltk.util import ngrams
n=1
sentence = corpus4
uni = ngrams(sentence.split(), n)
uni=list(uni)
dict=CountFrequency(uni)
pin = {}
for i in dict:
  t = i[0]
  pin[i[0]] = dict[i]
```

```python
import pandas as pd
df=pd.DataFrame(pin.items(),columns=['word','frequency'])
df
```

Top 10 most occuring words

```python
most_occuring=df.nlargest(n=10, columns=['frequency'], keep='all')
most_occuring
```
